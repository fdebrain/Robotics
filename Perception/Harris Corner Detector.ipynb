## Harris Corner Detecor

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as plt_img
import cv2
```

```python
class HarrisDetector:
    def __init__(self, path):
        self.candidates = np.empty(shape=(0,3)) # [x,y,score]
        self.corners = np.empty(shape=(0,2))
        
        # Load image
        img = cv2.imread(path)
        
        # Convert to grayscale (if necessary)
        if img.shape[2]>1:
            self.image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            self.image_gray = cv2.cvtColor(self.image, cv2.COLOR_RGB2GRAY)
        else:
            self.image_gray = img
            
        self.shape = self.image_gray.shape
        
        # Parameters
        self.window_size = 3
        self.kappa = 0.05
        self.threshold = 1e4
        self.nms_size = 10
    
    def blur(self, size, sigma=None):
        kernel = cv2.getGaussianKernel(size, sigma)
        self.image_blurred = cv2.filter2D(self.image_gray,-1,kernel)
    
    def compute_gradients(self):
        dy_kernel = np.array([-1,0,1])
        dx_kernel = np.reshape(dy_kernel, (1,-1))
        
        # Requires float pixel values (otherwise negative gradients are set to 0)
        self.Ix = cv2.filter2D(self.image_blurred.astype(np.float), -1, dx_kernel)
        self.Iy = cv2.filter2D(self.image_blurred.astype(np.float), -1, dy_kernel)
        
    def compute_cornerness(self):
        self.R = np.empty(self.shape)
        
        # Compute local structure matrix for each pixel
        N = self.window_size//2
        for i in range(N, self.shape[0] - N):
            for j in range(N, self.shape[1] - N):
                local_Ix = self.Ix[i-N:i+N,j-N:j+N]
                local_Iy = self.Iy[i-N:i+N,j-N:j+N]
                M = np.array([ [np.sum(local_Ix**2), np.sum(local_Ix*local_Iy)],
                               [np.sum(local_Ix*local_Iy), np.sum(local_Iy**2)]])
                
                self.R[i,j] = np.linalg.det(M) - self.kappa*(np.trace(M)**2)
                
                if self.R[i,j]>self.threshold:
                    self.candidates = np.vstack([self.candidates, np.array([i,j,self.R[i,j]])])
        
    def non_maxima_suppression(self):
        candidates_sorted = np.sort(self.candidates.view('i8,i8,f8'), order=['f2'], axis=0).view(np.float)[::-1]
        
        self.corners = np.vstack([self.corners, candidates_sorted[0,:2]])
        
        for candidate in candidates_sorted:
            # Compute distance between candidate and each corner
            dists = np.sqrt((candidate[0]-self.corners[:,0])**2 + (candidate[1]-self.corners[:,1])**2)

            # Test next candidate if close to any existing corner
            if (dists<self.nms_size).any():
                continue

            # Otherwise add current candidate to the list of corners
            self.corners = np.vstack([self.corners, candidate[:2]])
        
```

```python
#harris = HarrisDetector('images/test1.png')
#harris = HarrisDetector('images/demo.png')
harris = HarrisDetector('images/checkerboard.png')

harris.blur(3,1)
harris.compute_gradients()
harris.compute_cornerness()
harris.non_maxima_suppression()
```

```python
# Image and gradient images
fig, ax = plt.subplots(1,3,figsize=(20, 20));
ax[0].imshow(harris.image, cmap='gray');
ax[1].imshow(abs(harris.Ix), cmap='gray');
ax[2].imshow(abs(harris.Iy), cmap='gray');
```

```python
# Cornerness response
fig, ax = plt.subplots(figsize=(10, 10));
ax.imshow(harris.R, cmap='gray');
```

```python
# Image + detected corners (red dots)
fig, ax = plt.subplots(figsize=(10, 10));
ax.imshow(harris.image, cmap='gray');
ax.scatter(harris.corners[:,1], harris.corners[:,0], c='r');
```
